```python
import gradio as gr
import pandas as pd
import joblib

# Load the trained model
model = joblib.load("credit_fraud_detection_model.pkl")

# Define prediction function
def predict_fraud(transaction_type, amount, oldbalanceOrg, newbalanceOrig, oldbalanceDest, newbalanceDest):
    # Create DataFrame for input
    input_data = pd.DataFrame([{
        "type": transaction_type,
        "amount": amount,
        "oldbalanceOrg": oldbalanceOrg,
        "newbalanceOrig": newbalanceOrig,
        "oldbalanceDest": oldbalanceDest,
        "newbalanceDest": newbalanceDest
    }])

    # Predict
    prediction = model.predict(input_data)[0]
    if prediction == 1:
        return "🚨 Fraudulent Transaction Detected!"
    else:
        return "✅ Legitimate Transaction."

# Define Gradio interface
iface = gr.Interface(
    fn=predict_fraud,
    inputs=[
        gr.Dropdown(["PAYMENT", "TRANSFER", "CASH_OUT", "DEPOSIT"], label="Transaction Type"),
        gr.Number(label="Transaction Amount", value=1000.0),
        gr.Number(label="Old Balance (Sender)", value=10000.0),
        gr.Number(label="New Balance (Sender)", value=9000.0),
        gr.Number(label="Old Balance (Receiver)", value=0.0),
        gr.Number(label="New Balance (Receiver)", value=0.0),
    ],
    outputs=gr.Textbox(label="Prediction Result"),
    title="💳 Credit Card Fraud Detection",
    description="Enter transaction details to predict whether it’s fraudulent or legitimate.",
    theme="gradio/soft",
)

# Launch the app
iface.launch()

```
